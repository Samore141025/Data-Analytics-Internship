1.Explore the dataset and identify the number of rows and columns.

```python
# 1.Explore the dataset and identify the number of rows and columns.
# load the dataset
import pandas as pd
df = pd.read_csv("/content/sample_data/Dataset .csv")
```

2.Check for missing values in each column and
handle them accordingly.

```python
# check missing values
print("Missing values per column :")
print(df.isnull().sum())
```

3.Perform data type conversion if necessary.


```python
# check current data type
print(df.dtypes)
```

```python
# convert "has table booking" and "has online delivery" to binary (0/1)
df["Has Table booking"] = df["Has Table booking"].map({'Yes':1,'No':0})

df["Has Online delivery"] = df["Has Online delivery"].map({'Yes':1,'No':0})

print("\nUpdated data types :")
print(df[["Has Table booking","Has Online delivery"]].dtypes)
```

Analyze the distribution of the target variable
("Aggregate rating") and identify any class
imbalances

```python
import matplotlib.pyplot as plt
import seaborn as sns

# analyze the distribution of 'Aggregate rating'

plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()

# check for class imbalance

print("\nRating Value Counts:")
print(df['Aggregate rating'].value_counts().sort_index())
```
