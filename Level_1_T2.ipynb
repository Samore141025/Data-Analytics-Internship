Task: Descriptive Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
df = pd.read_csv("/content/sample_data/Dataset .csv")

# Preview
print(df.info())
df.head()
```

1.Calculate basic statistical measures (mean,
median, standard deviation, etc.) for numerical
columns.

```python
# list of key numerical columns
numerical_cols = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']

# get statistical summary
summary_stats = df[numerical_cols].describe()

# adding Median
medians = df[numerical_cols].median()

print("--- Basic Statistical Measures ---")
print(summary_stats)
print("\n--- Medians ---")
print(medians)
```

2.Explore the distribution of categorical
variables like "Country Code," "City," and
"Cuisines."

```python
# distribution for country code
print("Country Code Distribution:\n", df['Country Code'].value_counts())

# distribution for city
print("\nCity Distribution:\n", df['City'].value_counts().head(10))

# distribution for cuisines
print("\nCuisines Distribution:\n", df['Cuisines'].value_counts().head(10))
```

3.Identify the top cuisines and cities with the
highest number of restaurants

```python
# get the top 10 cities and cuisines
top_cities = df['City'].value_counts().head(10)
top_cuisines = df['Cuisines'].value_counts().head(10)

# top cities graph
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='viridis')
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

# top cuisines graph
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='magma')
plt.title('Top 10 Cuisines by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```
